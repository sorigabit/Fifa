```python
def sum_values(value_str):
    values = value_str.split('+')
    return sum(map(int, values))

columns_to_change = ['LS', 'ST', 'RS','LW', 'LF', 'CF', 'RF', 'RW', 'LAM','CAM', 'RAM', 'LM', 'LCM','CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB', 'LCB','CB', 'RCB', 'RB', 'GK']

fifa_data1[columns_to_change] = fifa_data1[columns_to_change].applymap(sum_values)
fifa_data1[columns_to_change] = fifa_data1[columns_to_change].astype(int)
```

```python
fifa_data1.head()
```

```python
fifa_data1['BP'] = fifa_data1['BP'].replace({'RB':'Defender', 'LB':'Defender', 'CB':'Defender', 'LWB' : 'Defender', 'RWB':'Defender'})
```

```python
fifa_data1['BP'] = fifa_data1['BP'].replace({'RW':'Midfielder', 'LW':'Midfielder', 'RM':'Midfielder', 'LM' : 'Midfielder', 'CDM':'Midfielder', 'CAM':'Midfielder','CM':'Midfielder'})
```

```python
fifa_data1['BP'] = fifa_data1['BP'].replace({'ST': 'Striker', 'CF': 'Striker'})
```

```python
fifa_data1['BP'].unique()
```

```python
def convert_currency (value):
    if 'K' in value:
        return float(value.replace('€', '').replace('K', '')) * 1000
    elif 'M' in value:
        return float(value.replace('€', '').replace('M', '')) * 1000000
    else:
        return float(value.replace('€', ''))
    
fifa_data1['Value'] = fifa_data1['Value'].apply(convert_currency)

print(fifa_data1['Value'])
```

```python
from sklearn.preprocessing import MinMaxScaler 
```

```python
y = fifa_data1['OVA']
X = fifa_data1.drop(['OVA'], axis=1)

X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = object)
```

```python
transformer = MinMaxScaler().fit(X_num)
X_normalized = transformer.transform(X_num)
X_normalized = pd.DataFrame(X_normalized,columns=X_num.columns)
display(X_normalized.head())
print(type(X_normalized))
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first').fit(X_cat)
print(encoder.categories_)
encoded = encoder.transform(X_cat).toarray()
print(encoded)

onehot_encoded = pd.DataFrame(encoded)
onehot_encoded.head(30)
```

```python
X = pd.concat([X_normalized, onehot_encoded], axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)
```

```python
import math
from sklearn import linear_model
```

```python
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
predictions = lm.predict(X_train)
r2_score(y_train, predictions)
```

```python
predictions_test = lm.predict(X_test)
r2_score(y_test, predictions_test)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,predictions_test)
mse
```

```python
rmse = np.sqrt(mean_squared_error(y_test,predictions_test))
rmse
```
